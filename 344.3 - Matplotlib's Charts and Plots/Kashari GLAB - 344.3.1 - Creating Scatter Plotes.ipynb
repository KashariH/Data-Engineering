### `Guided Lab - 34.3.1 - Example Scatter Plots` 

#### In this lab learners will learn how to visualize relationships between variables in their datasets effectively. 

```python
import matplotlib.pyplot as plt 
import numpy as np 

x = np.array([5, 7, 8, 7, 2, 17, 2, 9, 4, 11, 12, 9, 6]) # x- coordinates of the points 
y = np.array([99, 86, 87, 88, 111, 86, 103, 87, 94, 78, 77, 85, 86]) # y- coordinates of the points
plt.grid() 
plt.scatter(x, y, color = "red", s=90)  # Scatter plot with red color and size 90
plt.show() # Display the plot
# The above code creates a scatter plot using matplotlib.

# The c parameter specifies the colors of the markers based on the colors array. 
# The s parameter specifies the sizes of the markers based on the sizes array. 
```

#### Example 4: Combine color size & alpha. You can combine a colormap with different sizes of the dots. This is best visualized if the dots are transparent. Let's Create random arrays with 100 values for x-points, y-points, colors & sizes. 

```python
import matplotlib.pyplot as plt # Importing the matplotlib library for plotting 
import numpy as np # Importing the numpy library for numerical operations 

x = np.random.randint(100, size=(100)) # Generating 100 random integers between 0 and 100 for x- coordinates
y = np.random.randint(100, size=(100)) # Generating 100 random integers between 0 and 100 for y- coordinates
colors = np.random.randint(100, size=(100)) # Generating 100 random integers between 0 and 100 for colors
sizes = 10 * np.random.randint(100, size=(100)) # Generating 100 random integers between 0 and 100 for sizes

plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='nipy_spectral') # Scatter plot with colors and sizes based on the random arrays
plt.colorbar() # Adding a color bar to the plot
plt.show() # Display the plot
```

#### Example 5: Compare two Plots. 

#### The following example generates scatter plots to visualize age distribution across two different age groups: 20-39 years old and 40-69 years old. Each scatter plot represents random data points corresponding to the age group, with markers indicating the distribution of ages. The size of the markers represents the population density within each age group. The code utilizes NumPy to generate random data points and Matplotlib to create the scatter plots. 

```python
import numpy as np 
import matplotlib.pyplot as plt 

AgeGroupOne = range(40, 70, 1) # X-Values ranging from 40 to 69 year old
Purchase1=np.abs(np.random.randn(30) * 20) # Y-Values generated from a normal distribution 

plt.scatter(AgeGroupOne, Purchase1, # Scatter plot for ages 40-69
    s=200,  # Marker size
    alpha=0.6, # Transparency
    label='40-60 years old' # Label for the scatter plot 
    )

plt.scatter(AgeGroupOne, Purchase1, color= '#88c999') # Scatter plot for ages 40-69 with a different color

# Scatter plot for ages 20-39 
AgeGroupTwo = range(20, 40, 1) # X-Values ranging from 20 to 39 year old
Purchase2 = np.abs(np.random.randn(20) * 40) # Y-Values generated from a normal distribution
plt.scatter(AgeGroupTwo, Purchase2, 
    s=200, # Marker size
    alpha=0.6, # Transparency
    label='20-39 years old' # Label for the scatter plot
)

plt.scatter(AgeGroupTwo,Purchase2, color = '#88c984') 

# Display the plot 
plt.legend() # Adding a legend to the plot
plt.xlabel("Age") # X-axis label
plt.ylabel("Average purchase per weekend") # Y-axis label
plt.show() # Display the plot
```

#### Example 6: Randomly generated data points for scatter plot visualization 

#### The below example generates a scatter plot using randomly generated data points. It utilizes the NumPy library to generate random numbers and the Matplotlib library to create the plot.

#### The scatter plot represents the relationship between two sets of random data, where each data point is represented by a marker on the plot. The color and size of the markers are determined by additional randomly generated arrays, allowing for visual differentiation and enhancement. This code serves as a simple demonstration of how to create and customize scatter plots using Python libraries.

```python
import numpy as np # Importing the numpy library for numerical operations 
import matplotlib.pyplot as plt # Importing the matplotlib library for plotting

rng = np.random.RandomState(0) # Creates a random number generator-(RNG) object with a specific seed value of 0. This ensures reproducibility of the random numbers generated. 
x = rng.randn(100) # Generates an array x of the 100 random numbers drawn from a standard normal distribution (mean=0, variance=1). 
y = rng.randn(100) # Generates an array y of the 100 random numbers drawn from a standard normal distribution (mean=0, variance=1).

colors = rng.rand(100) # Generates an array of 100 random numbers between 0 and 1 for colors 
sizes = 1000 * rng.rand(100) # Generates an array of 100 random numbers between 0 and 1 for sizes, multiplied by 1000 to increase the size of the markers

plt.scatter(x, y, c=colors, s=sizes, alpha=0.3) # Creates a scatter plot with x and y coordinates, colors, sizes, and transparency (alpha=0.3)
plt.title("Scatter chart") # Title of the plot
plt.xlabel("Random data for x") # X-axis label
plt.ylabel("Random data for y") # Y-axis label

# The cmap parameter specifies the colormap to be used for the colors.
plt.colorbar() 
plt.show() 
```

#### Example 7: Real Time Data analysis and Data Visualization 

#### Now lets analyse on a realtime data. So in the code below, I will use the housing prices and I will plot the data to analyze the median prices and the population. The idea is to visualize the densely populated areas with bigger circles and the areas with high prices with darker circles and vice versa:

```python
import pandas as pd 
housing_df = pd.read_csv("https://raw.githubusercontent.com/ageron/handson-ml/master/datasets/housing/housing.csv") # Load the housing dataset from a CSV file
housing_df # Display the first few rows of the dataset
```

```python
plt.figure(figsize=(12, 8)) # Create a new figure with a specified size of 12x8 inches
plt.scatter(
    x=housing_df['longitude'], #X-Values for longitude 
    y=housing_df['latitude'], # Y-Values for latitude
    alpha=0.4, # Transparency of the markers
    s=housing_df['population'] / 100, # Size of the markers based on population
    c=housing_df['median_house_value'], # Color of the markers based on median house value 
    cmap=plt.get_cmap('jet') # Colormap to be used for the colors 
)

# c='median_house_value' specifies that the color of the markers based on the 'median house value' column. However, this parameter expects a sequence of color values or a colormap, not a column name. 
plt.colorbar(label='Median House Value') # Adding a color bar to the plot with a label
plt.legend(['Population']) # Adding a legend to the plot 
plt.xlabel('Longitude') # X-axis label
plt.ylabel('Latitude') # Y-axis label 
plt.title('Housing Data Scatter Plot') # Plot title
plt.show() # Display the plot
```
