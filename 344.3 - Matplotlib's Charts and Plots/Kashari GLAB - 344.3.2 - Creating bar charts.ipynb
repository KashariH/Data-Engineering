### `Guided Lab - 344.3.2 - Creating Bar Charts`

#### In this lab learners will learn how to create bar charts using Matplotlib to visualize categorical data effectively. 

##### Example 1: Create Simple Bar Charts Vertically 

```python
import numpy as np 
import matplotlib.pyplot as plt

CoursesName = ['C Course', 'C++ Course', 'Java Course', 'Python Course'] # Assigning x and y coordinates 
Student = [90, 60, 130, 190] 

plt.bar(CoursesName, Student, color='green', width = 0.3) # Creating bar chart 
plt.xlabel('Courses Name') # Adding x-axis label 
plt.ylabel('Student Registrations number') # Adding y-axis label

# Displaying the title of the bar chart
plt.title('Number of Students in Different Courses')
plt.show() # Displaying the bar chart
```

##### Example 2: Create Simple Bar Charts Horizontally 

```python
import numpy as np 
import matplotlib.pyplot as plt 

CoursesName = ['C Course', 'C++ Course', 'Java Course', 'Python Course'] # Assigning x and y coordinates
Student = [90, 60, 130, 190]

plt.barh(CoursesName, Student, color='green', height = 0.3) # For horizontal bars, use height instead of width. # Plotting the bar chart.
plt.xlabel('Student Registrations number') # Adding x-axis label 
plt.ylabel('Courses Name') # Adding y-axis label

plt.title('Number of Students in Different Courses') # Displaying the title of the bar chart
plt.show() # Displaying the bar chart
```

##### Example 3.1: Matplotlib Bars - Stacked Bar Charts 

##### A stacked bar chart is a type of graph that displays multiple data points on top of each other. In a stacked bar chart, each bar represents a single category that contains smaller categories. Stacked bar chart plots need to provide the additional bottom parameter that indicate where the bar should start. 

```python
import numpy as np 
import matplotlib.pyplot as plt 

CoursesName = ['C Course', 'C++ Course', 'Java Course', 'Python Course'] # Assigning x and y coordinates
StudentGroup1 = [90, 60, 130, 600] # Assigning x coordinates
StudentGroup2 = [100, 30, 160, 690] # Assigning y coordinates
width = 0.3 # Setting the width of the bars 

plt.bar(CoursesName, StudentGroup1, color='green', width = 0.3) # Creating the first bar chart
plt.bar(CoursesName, StudentGroup2, bottom=StudentGroup1, width = 0.3) # Creating the second bar chart

plt.xlabel('Courses Name') 
plt.ylabel('Student Registrations number') # Adding y-axis label
plt.title('Number of Students in Different Courses') # Displaying the title of the bar chart

plt.show() # Displaying the bar chart 
```

##### Example 3.2: Lets see one more example for a stacked Bar plot

```python
years = ['2016', '2017', '2018', '2019'] # Assigning x and y coordinates
rice_price = [57, 50, 47, 30] # Assigning x coordinates
wheat_price = [43, 50, 53, 70] # Assigning y coordinates

plt.figure(figsize=(10, 6)) # Setting the figure size


plt.bar(years, rice_price, label = 'Rice') # Creating the first bar chart
plt.bar(years, wheat_price, bottom=rice_price, label = 'wheat') # Creating the second bar chart

plt.xlabel('Years', fontsize=14) # Adding x-axis label
plt.ylabel('Preference (%)', fontsize=14) # Adding y-axis label
plt.title('The Price of Rice and Wheat', fontsize=20) # Displaying the title of the bar chart
plt.legend() # Adding a legend to the chart
plt.show() # Displaying the bar chart
```

##### Example 3.3

```python
years = ['2016', '2017', '2018', '2019'] # Assigning x and y coordinates
rice = np.array([57, 50, 47, 30]) # Assigning x coordinates
wheat = np.array([43, 50, 53, 70]) # Assigning y coordinates
mustard = np.array([10, 16, 13, 15]) # Assigning y coordinates

plt.figure(figsize=(10, 6)) # Setting the figure size

plt.bar(years, rice, label='rice') # Creating the first bar chart
plt.bar(years, wheat, bottom=rice, label='Wheat') # Creating the second bar chart
plt.bar(years, mustard, bottom=rice+wheat, label='Mustard seeds') # Creating the third bar chart

plt.xlabel('Years', fontsize=14) # Adding x-axis label
plt.ylabel('Preference (%)', fontsize=14) # Adding y-axis label
plt.title('The results of Rice,wheat and mustard seeds export', fontsize=20) # Displaying the title of the bar chart
plt.legend() # Adding a legend to the chart
plt.show() # Displaying the bar chart
```

##### Example 4: Compare Plots - Group Bar Charts in a single figure 

```python
import numpy as np 
import matplotlib.pyplot as plt 

CoursesName = ['C Course', 'C++ Course', 'Java Course', 'Python Course']
StudentGroup1 = [90, 60, 130, 600]
StudentGroup2 = [100, 30, 160, 690]
width = 0.3

x = np.arange(len(CoursesName)) # Calculate the position for the bars 

plt.bar(x - width/2, StudentGroup1, color='red', width=width, label='Student Group 1') # Plotting
plt.bar(x + width/2, StudentGroup2, color='green', width=width, label='Student Group 2')

plt.xlabel('Courses Name')
plt.ylabel('Number of Students')

plt.title('Number of Students in Different Courses')

plt.xticks(x, CoursesName, rotation=45, ha='right') # Rotate x-axis labels for better readability 

plt.legend() # Adding a legend to the chart 
plt.show() # Displaying the bar chart
```

##### Example 5: Creating a Bar Chart using Pandas and Matplotlib 

```python
import pandas as pd 
import matplotlib.pyplot as plt 

data = {'CoursesName': ['C programming Course', 'C++ Programming Course', 'Java Programming Course', 'Python Programming Course'], 'Student': [90, 60, 130, 600]}
df = pd.DataFrame(data) # Creating a DataFrame from the data dictionary 

plt.bar(df['CoursesName'], df['Student'], color='green')
# Plotting the bar chart using DataFrame columns
plt.xlabel('Courses Name')
plt.ylabel('Number of Students')

plt.title('Number of Students in Different Courses')

plt.xticks(rotation=45, ha='right') # Rotate x-axis labels for better readability 
plt.show() # Displaying the bar chart 
```

##### Example 6: Creating a Group Bar Chart Using Pandas and Matplotlib 

```python
import pandas as pd
import matplotlib.pyplot as plt 

data = {'CoursesName': ['C Programming Course', 'C++ Programming Course', 'Java Programming Course', 'Python Programming Course'],
        'Student Group 1': [90, 60, 130, 600],
        'Student Group 2': [100, 30, 160, 690]} # Sample data 

df = pd.DataFrame(data) # Creating a DataFrame from the data dictionary 

bar_width = 0.35
index = df.index

plt.bar(index - bar_width/2, df['Student Group 1'], bar_width, label='Student Group 1', color='blue') # Plotting the first group of students
plt.bar(index + bar_width/2, df['Student Group 2'], bar_width, label='Student Group 2', color='orange') # Plotting the second group of students

plt.xlabel('Courses Name')
plt.ylabel('Number of Students')

plt.title('Number of Students in Different Courses')

plt.xticks(index, df['CoursesName'], rotation=45, ha='right') # Rotate x-axis labels for better readability

plt.legend() # Adding a legend to the chart 

plt.tight_layout() # Adjust layout to prevent overlap 
plt.show() # Displaying the bar chart 
 
```
