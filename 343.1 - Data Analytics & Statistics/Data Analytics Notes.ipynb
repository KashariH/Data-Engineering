# `Data Analytics Notes`

#### Data analysis is the process of gathering, systemically examining, cleansing, transforming, analyzing, mining, and interpreting data to discover meaningful patterns, insights, and trends.

![Screenshot 2025-05-07 132435.png](<attachment:Screenshot 2025-05-07 132435.png>)

#### By utilizing various techniques and tools, data analytics helps to extract valuable information from raw data. It enables informed decision-making and provides a deeper understanding of complex datasets. 

### `Data Analytics Vs. Data Analysis`

#### Data Analysis: Detailed examination of the elements or structure of something (Can be done without numbers or data such as business analysis pyscho analysis, etc). 

#### Data Analytics: The systematic computational analysis of data or statistics (Even when used without the prefix "Data" almost invariably implies the use of data for performing numerical manipulation and inference).  

#### Experts say that Data Analysis is based on inferences based on historical data. Whereas, Data Analytics is for predicting future performance. 

### `Data Analytics or Data Scientist`

#### Includes data analysis, plus predictive, prescriptive, and automated decision-making techniques. 

#### Encompassing data analysis, however includes advanced techniques like predictive modeling, machine learning, and statistical analysis. 

#### May build predictive models, automate analysis pipelines, or integrate data from multiple sources

#### Aimed at forecasting future trends or optimizing processes ("What will happen"), or ("How can we make it happen"). 

#### Example: Using customer data to build a model that predicts churn and recommends retention strategies. 

### `Data Analysis or Data Analyst`

#### Interpreting data for insights. 

#### Exploring datasets to answer specific questions ("What Happened"), or ("Why did it happen"). Uses tools like Excel, SQL, Python, or Tableau to generate reports, dashboards, and insights. 

#### Example: Analyzing customer purchase data to determine which products are selling best in Q1. 

### `Four Types Of Data Analysis` 

#### Descriptive Analysis - What Happened? (Summarizes past data to understand changes or trends). For example: Monthly sales reports, average customer ratings, website traffic statistics. 

#### Diagnostic Analysis - Tells us why something happened. 

#### Predictive Analytics - Tells us what will likely happen in the future. 

#### Prescriptive Analytics - Tells us how to act

### `Data Mining`

#### Data Mining is a type of data analysis technique used to carry out data modeling, as well as knowledge discovery that is geared towards predictive purposes. Data Mining is the analysis of large quantities of data to extract previously unknown, interesting patterns of data, unusual data, its dependencies. 

#### The goal is the extraction of patterns and knowledge from large amounts of data and not the extraction of data itself. 

### `Data Wrangling` 

#### Data wrangling, also known as data munging is an iterative process that involves data exploration, transformation, validation, and making it available for a credible and meaningful analysis

#### Includes a range of tasks involved in preparing raw data for a clearly defined purpose, where raw data at this stage is data that has been collated through various data sources in a data repository. 

#### Data Wrangling captures a range of tasks involved in preparing data for analysis. Typically, it is a four-step process that involves Discovery, Transformation, Validation, and Publishing.
