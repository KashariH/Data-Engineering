# `Python Integration With SQL`

## Learning Objective : I will be able to develop a straightforward Python application to interact with MySQL databases.

#### The mysql-connector-python is a python driver (or connector) for MySQL databases. It allows python developers to connect to, and interact with MySQL database systems from within their python applications. In order for MySQL to interact with Python you have have to I will need to install a Python SQL driver for MySQL 

#### (mysql-connector-python) is a MySQL Connector for Python. you can download this module with pip. 

#### Python can raise an exception. An exception can occur if MySQL is not available, or if the database does not exist or is an invalid username or password. You can check that the connection to the MySQL database has been established successfully by using the is_connected() method. 

#### A parameterized query is a query in which placeholders (%s) are used for parameters and the parameter values are supplied at execution time. We will use (Parameterized query) to use Python variable in SQL query. 

# `Inserting Data into MySQL Table From Python`

#### For single row: You will use the execute() method of cursor object to insert a SINGLE RECORD. 

#### For multiple row: You will use the executemany() method of cursor object to insert MULTIPLE  RECORDS into a table. 

#### You have to explicitly call the commit() method in order to make the changes to the database. 

#### Get the number of rows affected: After a successful insert operation, use a cursor.rowcount() method to get the number of rows affected. The count depends on how many rows you are inserting.
