##### Example 1 - Seaborn Scatter Plots 

##### Scatter plots are used to visualize the relationship between two continuous variables. Each point on the plot represents a single data point, and the position of the point on the x and y-axis represents the values of the two variables. 

```python
import seaborn as sns # imported seaborn as sns 
import matplotlib.pyplot as plt # imported matplotlib.pyplot as plt
tips = sns.load_dataset("tips") # loaded the tips dataset from seaborn
tips.head(10) # displayed the first 10 rows of the tips dataset
```

```python
sns.scatterplot(x="total_bill", y="tip", hue="sex", size="size", sizes=(50, 200), data=tips) # The hue parameter is used to add a categorical variable to the plot by coloring the points based on different categories. # The size parameter is used to vary the size of the points based on another variable, in this case, the size of the party. The sizes parameter specifies the range of sizes for the points.
plt.xlabel("Total Bill") # The x-axis label is set to "Total Bill"
plt.ylabel("Tip") # The y-axis label is set to "Tip"
plt.title("Relationship between Total Bill and Tip") # The title of the plot is set to "Relationship between Total Bill and Tip"
plt.show() # displayed the plot
```

##### Example 2 - Seaborn Line Plots

##### Line plots are used to visualize trends in data over time or other continuous variables. In line plot, each data is connected by a line, creating a smooth curve. 

```python
import seaborn as sns  # imported seaborn as sns

fmri = sns.load_dataset("fmri") # loaded the fmri dataset from seaborn
fmri.head(5) # displayed the first 5 rows of the fmri dataset

sns.lineplot(x="timepoint", y="signal", sizes=(50, 200), data=fmri) # The sizes parameter specifies the range of sizes for the points.
```

```python
import seaborn as sns 
import matplotlib.pyplot as plt 

fmri = sns.load_dataset("fmri") 

sns.lineplot(x="timepoint", y="signal", hue="event", style="region", markers=True, dashes=False, sizes=(70, 200), data=fmri) # The markers parameter is set to True to display different markers for different levels of the "region" variable. The dashes parameter is set to False to use solid lines instead of dashed lines.

plt.xlabel("Timepoint") # The x-axis label is set to "Timepoint"
plt.ylabel("Signal Intensity") # The y-axis label is set to "Signal Intensity" 
plt.title("Changes in Signal Intensity over Time") # The title of the plot is set to "Changes in Signal Intensity over Time"
plt.show() # displayed the plot 
```

##### Vertical Lines with axvline() Functions. If you want to add a vertical line to a plot made with Seaborn, you can use the axvline function and set where you want to display the vertical line with the ( x argument) 

```python
ax = sns.lineplot(x="timepoint", y="signal", color = "purple", alpha = 0.1, data=fmri) # The alpha parameter is set to 0.1 to make the lines more transparent.
ax.axvline(x=5) # added a vertical line at x=5 
plt.xlabel("Timepoint") # The x-axis label is set to "Timepoint" 
plt.ylabel("Signal Intensity") # The y-axis label is set to "Signal Intensity" 
plt.title("Changes in Signal Intensity over Time") # The title of the plot is set to "Changes in Signal Intensity over Time"
```

##### Horizontal lines with axhline() functions. Similarly, the axhline() function allows adding horizontal lines to the Seaborn plots, but now you need to set the value with the y argument. 

```python
ax = sns.lineplot(x="timepoint", y="signal", color = "purple", alpha = 0.1, data=fmri) # The color parameter is set to "purple" to use a combination of colors for the lines. 
ax.axvline(x = 5) # added a vertical line at x=5 
ax.axhline(y = 0.17,  linestyle = "dashed" ) # added a horizontal dashed line at y=0.17 

plt.xlabel("Timepoint") # The x-axis label is set to "Timepoint"
plt.ylabel("Signal Intensity") # The y-axis label is set to "Signal Intensity" 
plt.title("C I S I O T") # The title of the plot is set to "C I S I O T"

```

##### Example 3 - Seaborn bar plots. Bar plots are used to visualize the relationship between a categorical variable and a continuous variable. In bar plot, each bar represents the mean or median ( or any aggregation) of the continuous variable for each category. 

```python
import seaborn as sns

titanic = sns.load_dataset("titanic") # loaded the titanic dataset from seaborn
titanic.head(5) # displayed the first 5 rows of the titanic dataset

sns.barplot(x="class", y="fare", data=titanic)
```

```python
import seaborn as sns

import matplotlib.pyplot as plt
titanic = sns.load_dataset("titanic")

sns.barplot(x="class", y="fare", hue="sex", data=titanic) # The hue parameter is used to add a categorical variable to the plot by coloring the bars based on different categories.
plt.xlabel("Class") # The x-axis label is set to "Class"
plt.ylabel("Fare") # The y-axis label is set to "Fare"
plt.title("Average Fare by Class and Gender on the Titanic") # The title of the plot is set to "Average Fare by Class and Gender on the Titanic" 
plt.show() # displayed the plot
```

##### Example 4 - Seaborn histograms. Histograms visualize the distribution of a continuous variable. In a histogram, divided into bins and the height of each bin represents the frequency or count of data points within that bin. 

```python
import seaborn as sns 

iris = sns.load_dataset("iris") # loaded the iris dataset from seaborn 
iris.head(10) # displayed the first 10 rows of the iris dataset 

sns.histplot(x="petal_length", data=iris) # The histplot function is used to create a histogram of the "petal_length" variable in the iris dataset.
```

```python
import seaborn as sns

import matplotlib.pyplot as plt # imported matplotlib.pyplot as plt 


iris = sns.load_dataset("iris") # loaded the iris dataset from seaborn 

sns.histplot(data=iris, x="petal_length", bins=20, kde=True, color="green") # The bins parameter is set to 20 to specify the number of bins in the histogram. The kde parameter is set to True to overlay a kernel density estimate on top of the histogram. The color parameter is set to "green" to change the color of the histogram bars.

plt.xlabel("Petal Length (cm)") # The x-axis label is set to "Petal Length (cm)" 

plt.ylabel("Frequency") # The y-axis label is set to "Frequency" 

plt.title("Distribution of Petal Length in Iris Dataset") # The title of the plot is set to "Distribution of Petal Length in Iris Dataset"
plt.show() # displayed the plot 
```

##### Example 5 - Seaborn density plots. Density plots also known as kernel density plots, are a type of data visualization that display the distribution of a continuous variable. They are similar to histograms, but instead of representing the data as bars, density plots use a smooth curve to estimate the density of the data. 

```python
import seaborn as sns 

tips = sns.load_dataset("tips") # loaded the tips dataset from seaborn

sns.kdeplot(data=tips, x="total_bill") # The kdeplot function is used to create a kernel density estimate plot of the "total_bill" variable in the tips dataset.
```

##### We can use the hue parameter to differentiate between "lunch" and "dinner" meal times. We use fill parameter to fill the area under the curve. We adjust the alpha and linewidth parameters to make the plot more visually appealing. 

```python
import seaborn as sns

import matplotlib.pyplot as plt

tips = sns.load_dataset("tips") # loaded the tips dataset from seaborn

sns.kdeplot(data=tips, x="total_bill", hue="time", fill=True, alpha=0.5, linewidth=1.5) # The fill parameter is set to True to fill the area under the curve. The alpha parameter is set to 0.5 to make the filled area semi-transparent. The linewidth parameter is set to 1.5 to increase the thickness of the lines. 

plt.title("Density Plot of Total Bill by Meal Time") # The title of the plot is set to "Density Plot of Total Bill by Time"
plt.xlabel("Total Bill") # The x-axis label is set to "Total Bill" 
plt.ylabel("Density") # The y-axis label is set to "Density" 
plt.show() # displayed the plot 
 
```
