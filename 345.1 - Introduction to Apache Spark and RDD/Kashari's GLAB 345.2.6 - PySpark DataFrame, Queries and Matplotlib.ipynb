### `Guided Lab 345.2.6 - PySpark DataFrame, Queries and Matplotlib`

##### In this lab we will demonstrate and apply Python Pandas 
##### Learners will be able to analyze the sales dataset using PySpark and Matplotlib 
##### Save data into RDBMS 

```python
# Load / Read Data from the CSV file
import pyspark
from pyspark.sql import SparkSession

# create the SparkSession
spark = SparkSession.builder.appName('DemoApp').getOrCreate()

df = spark.read.load("C:/Users/Sales_October_2019.csv", format="csv", header = True,inferSchema = True)


```

```python
# Inspecting the dataset by using the below lines 
df.printSchema()
df.columns
df.dtypes
df.head()
df.show(truncate=False)

```

##### Example 1.1 - Get the sum of price and quantity column 

```python
df.select("Product","Price Each","Quantity Ordered","Order Date").groupby("Product").sum().show()

```

##### Example 1.2 - Get the sum of the cost column grouped by item. 

```python
df.groupBy('Product').sum('Price Each').show()

```

##### Example 2: Matplotlib and Pyspark 
##### The pyspark DataFrame can be converted to a python pandas DataFrame using a function topandas(). 

##### Example 2.1 - Using toPandas(). 

```python
df_sample_pandas = df.toPandas()
df_sample_pandas.hist(bins=300, figsize=(12,8),  legend=True)



```

##### Example 2.2 - using toPandas() 

```python
import matplotlib.pyplot as plt
roc = df.toPandas()
plt.plot(roc['Quantity Ordered'], roc['Order ID'])

plt.ylabel('Qrder ID')
plt.xlabel('Quantity Ordered')
plt.title(' order id and quantity')
plt.show()

```

##### Example 2.3 - Write and save data into SQL 

##### Classicmodels is the database name and salesdata is the table name, but you can change the database and table name. The below lines will write/store all DataFrame (column) data into the database.


```python
df.write.format("jdbc") \
  .mode("append") \
  .option("url", "jdbc:mysql://localhost:3306/classicmodels") \
  .option("dbtable", "classicmodels.salesdata") \
  .option("user", "root") \
  .option("password", "password") \
  .save()

```

##### We can write/store a specific DataFrame (column) in the database by using the select() method, as shown below:


```python
df.select("Order ID","Product").write.format("jdbc") \
  .mode("append") \
  .option("url", "jdbc:mysql://localhost:3306/classicmodels") \
  .option("dbtable", "classicmodels.salesdata") \
  .option("user", "root") \
  .option("password", "password") \
.option("header","false")\
  .save()

# Only Order ID and Product data will be saved in the database. 
```
