## `What is an API?` An application programming Interface (API) defines the rules that you must follow to communicate with other software systems. Developers expose or create APIs so that other applications can communicate with their applications programmatically. 

## Think of an API as A mediator between the (users) or (clients) and the resources or web services they want to get. It is also a way for an organization to share resources and information while maintaining security, control, and authentication -- determining who gets access to what. 

## Rest web services is any web service that adheres to REST architecture constraints. These web services expose their data to the outside world through an API. REST APIs provide access to web service data through public web URLs. 

### 

## The basic function of a RESTful API is the same as browsing the internet. 

## `HTTP METHODS` - Used to define the action that the client wants to perform on the resource. 

## There are eight standard HTTP methods: GET (Retrieve a resource), POST (Create a new resource), PUT(Update an existing resource), DELETE(Delete a resource), HEAD(Retrieve the headers of a resource), OPTIONS(Query the capabilities of the resource), TRACE(Echo back the request message), CONNECT(Establish a tunnel to a remote server) 

## `GET VS POST`

## (`POST`): Will contain all of the data necessary for creating a new object. Using the post(), we can perform changes to the server. Used for defining a new piece of data. 

## (`GET`): Used for reading something. Using get(), data is retrieved from the server. The GET request is used for identifying which piece of data the client would like to read. 

## `PUT VS PATCH`

## PUT & PATCH ARE BOTH USED FOR UPDATING. WHENEVER YOU UPDATE YOUR FACEBOOK PROFILE YOU ARE MAKING A PUT OR PATCH REQUEST. 

## (`PUT`): would be used to completely replace your profile. = Replacing all of the data

## `(PATCH)`: Used to change a few fields of your profile. = Replacing part of the data
