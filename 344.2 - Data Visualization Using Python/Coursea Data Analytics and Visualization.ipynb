#### `Coursea Notes On Data Visualization & Analytics`

#### Data Visualization is a great tool for making the connection between an image and the information it represents. 

##### The venn-diagram below shows us the four elements that are needed for a successful visualization. (Information, Story, Goal, Visual Form)

##### You can also, use these questions to determine if your data is effective: What is the practical question?, What does the data say?, What does the visual say?

#### `Bar Graphs`: Use size contrast to compare two or more values. 
#### `The X-axis` (Horizontal Line) is used to represent time periods, categories, or other variables. 
#### `The Y-axis` (Vertical Line) is used scale of values for the variables 
#### Bar graphs are a great way to clarify trends. 


#### `Line Graphs`: Help your audience understand shifts or changes in your data
#### Usually used to track changes through a period of time, but they can be paired with other factors as well. 

#### `Pie Charts`: Shows how much each part of something makes up the whole

#### `Maps`: Help organize data geographically
